```python
import ipywidgets as widgets
from IPython.display import display
import time

# --- Tab 1: Input ---
url_input = widgets.Text(placeholder="Paste shapefile URL", description="URL:", layout=widgets.Layout(width='400px'))
run_button = widgets.Button(description="Run Model", button_style='success')
input_tab = widgets.VBox([url_input, run_button])

# --- Tab 2: Logs ---
log_output = widgets.Output()
log_tab = widgets.VBox([log_output])

# --- Tab 3: Output ---
result_output = widgets.Output()
output_tab = widgets.VBox([result_output])

# --- Wire up button ---
def on_run(b):
    with log_output:
        for i in range(5):
            print(f"Step {i+1}: doing something...")
            time.sleep(0.5)
    with result_output:
        print(f"Done! URL was: {url_input.value}")

run_button.on_click(on_run)

# --- Assemble tabs ---
tabs = widgets.Tab(children=[input_tab, log_tab, output_tab])
tabs.set_title(0, "Input")
tabs.set_title(1, "Logs")
tabs.set_title(2, "Output")

display(tabs)
```
